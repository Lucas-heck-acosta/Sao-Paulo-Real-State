```python
df_rent = df_rent.drop_duplicates()
```

```python
Y = df_rent['Price']
X = df_rent.drop(['Price'], axis=1)
```

```python
X.head(1)
```

```python
X = X.drop(['Negotiation Type', 'District', 'Property Type'], axis = 1)
```

```python
X.head(1)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X[0]
```

```python
X_training, X_test, Y_training, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
X_training.shape, Y_training.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_bayer = GaussianNB()
naive_bayer.fit(X_training, Y_training)
```

```python
predict = naive_bayer.predict(X_test)
```

```python
accuracy_score(predict, Y_test)
```
